```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
pd.pandas.set_option('display.max_columns',None)
pd.pandas.set_option('display.max_rows',None)
```

```python
df=pd.read_csv('zomato.csv',encoding='latin-1')
df.head(10000)
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head(20)
```

```python
df2=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
df2.head(1000)
```

```python
df2.dtypes
```

```python
countries=df2.Country.value_counts().index
```

```python
df2.Country.value_counts().values
```

```python
plt.figure(figsize=(12,6))
plt.pie(df2.Country.value_counts().values[:3],labels=countries[:3],autopct='%1.2f%%')
```

Observation : Maximum transactions of Zomato are from India, after that in US and then in UK

```python
df2.columns
```

```python
df_rating=df2.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
df_rating
```

So, ratings goes like
1. 4.5 - 4.9 = Excellent
2. 4.0 - 4.4 = Very good
3. 3.5 - 3.9 = Good
4. 2.5 - 3.4 = Average
5. 1 - 2.4 = Poor
6. 0 = not rates

```python
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x=df_rating['Aggregate rating'],y=df_rating['Rating count'])
```

```python
# we can use rating color in our bar plot
sns.barplot(x=df_rating['Aggregate rating'],y=df_rating['Rating count'],hue=df_rating['Rating color'],palette=['white','red','orange','yellow','green','green'])
```

Not rated count is very high.
Graph is in the form of guassian distribution.
maximum ratings are from 2.8 to 4.1

```python
# Countries with zero ratings
df2[df2["Aggregate rating"]==0.0].groupby('Country').size().reset_index()
```

maximum ratings with 0 are from India.

```python
df2.groupby(['Currency','Country']).size().sort_values().reset_index()
```

```python
df2[df2['Has Online delivery']=='Yes'].groupby('Country').size()
```
